### CHEOPS timing test target S/N calculation

Taking S/N/minute values from http://cheops.unibe.ch/science/measurement-requirements/ as follows

- V=12,1100 ppm/min
- V=9, 150 ppm/min
- V=6, use 50 ppm in 20 minutes systematic noise requirement

Assume that we want to achieve a timing precision is $\pm$ 1 second in an observing sequence of 12 hours.

Use $\sigma(\phi) = \sqrt{\frac{2}{N}}\frac{\sigma(m)}{a}$ from Montgomery, M. H., & Oâ€™Donoghue, D. 1999, Delta Scuti Star Newsletter, 13, 28 to calculate amplitude required to get 1 second timing precision as a function of pulsation frequency for all 3 cases. 

```python
%pylab inline
```

```python
frequency =  linspace(25,2500)  # microHz

# V = 12, one observation per minute
sigma_m= 1100e-6
N = 12*60
A_12 = sqrt(2/N)*sigma_m/(frequency*1e-6)

# V = 9, one observation per minute
sigma_m= 150e-6
N = 12*60
A_9 = sqrt(2/N)*sigma_m/(frequency*1e-6)

# V = 6, one observation per 20 minutes
sigma_m= 50e-6
N = 12*60/20
A_6 = sqrt(2/N)*sigma_m/(frequency*1e-6)
```

```python
loglog(frequency,A_12,label='V=12')
loglog(frequency,A_9,label='V=9')
loglog(frequency,A_6,label='V=6')
xlabel("Frequency [$\mu$Hz]")
ylabel("Amplitude")
legend()
```
